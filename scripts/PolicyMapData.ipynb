```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing each dataset into it's own dataframe
data = pd.read_csv("HomeOwners.csv", na_values = '?')
data2 = pd.read_csv("MedianHomeValue.csv", na_values = '?')
data3 = pd.read_csv("MedianHousholdIncome.csv", na_values = '?')
data4 = pd.read_csv("NumberOfRenters.csv", na_values = '?')
data5 = pd.read_csv("VacancyRate.csv", na_values = '?')
data.head()
```

```python
#Dropping first row of data from each dataframe as it was redundant with column names
data.drop(index=0, inplace=True)
data2.drop(index=0, inplace=True)
data3.drop(index=0, inplace=True)
data4.drop(index=0, inplace=True)
data5.drop(index=0, inplace=True)

#Checking Datatypes for each dataframe
data2.info()
```

```python
data3.info()
```

```python
data4.info()
```

```python
data5.info()
```

```python
# replace spaces with underscores in column names
data.columns = data.columns.str.replace(' ', '_')
data2.columns = data2.columns.str.replace(' ', '_')
data3.columns = data3.columns.str.replace(' ', '_')
data4.columns = data4.columns.str.replace(' ', '_')
data5.columns = data5.columns.str.replace(' ', '_')

# change datatypes of the columns I'm going to keep
data['GeoID'] = data['GeoID'].astype(int)
data['Homeownership_Rate'] = data['Homeownership_Rate'].astype(float)

data2['GeoID'] = data2['GeoID'].astype(int)
data2['Median_Home_Value'] = data2['Median_Home_Value'].astype(float)

data3['GeoID'] = data3['GeoID'].astype(int)
data3['Median_Household_Income'] = data3['Median_Household_Income'].astype(float)

data4['GeoID'] = data4['GeoID'].astype(int)
data4['Rental_Occupancy_Rate'] = data4['Rental_Occupancy_Rate'].astype(float)

data5['GeoID'] = data5['GeoID'].astype(int)
data5['Vacancy_Rate'] = data5['Vacancy_Rate'].astype(float)




data.info()
```

```python
# keeping only the columns I'm going to use
dataHomeOwner = data[['GeoID', 'Homeownership_Rate']]
dataMHV = data2[['GeoID', 'Median_Home_Value']]
dataMHI = data3[['GeoID', 'Median_Household_Income']]
dataROR = data4[['GeoID', 'Rental_Occupancy_Rate']]
dataVR = data5[['GeoID', 'Vacancy_Rate']]

dataHomeOwner.head()
```

```python
# merging columns into one dataframe
data_merged = pd.merge(dataHomeOwner, dataMHV, on='GeoID')
data_merged.head()
```

```python
# merging columns into one dataframe
data_merged = pd.merge(data_merged, dataMHI, on='GeoID')
data_merged.head()
```

```python
# merging columns into one dataframe
data_merged = pd.merge(data_merged, dataROR, on='GeoID')
data_merged.head()
```

```python
# merging columns into one dataframe
data_merged = pd.merge(data_merged, dataVR, on='GeoID')
data_merged.head()
```

```python
# Checking for missing values
rows_with_nan = data_merged[data_merged.isnull().any(axis=1)]
rows_with_nan.head()
```

```python
# Deleting rows with no data at all
df_multiple_deleted = data_merged.drop(data_merged.index[[19, 62, 97]])
```

```python
# checking remaining nan values
rows_with_nan = df_multiple_deleted[df_multiple_deleted.isnull().any(axis=1)]
rows_with_nan.head()
```

```python
# calculate median column value for the remaining nan
mhv_median = data_merged['Median_Home_Value'].median()
```

```python
# replacing the nan with the median value
data_merged['Median_Home_Value'] = data_merged['Median_Home_Value'].fillna(mhv_median)
```

```python
# Removing the blank rows from the dataframe
data_merged = data_merged.drop(data_merged.index[[19, 62, 97]])
```

```python
# Verifying no more blank values
rows_with_nan = data_merged[data_merged.isnull().any(axis=1)]
rows_with_nan.head()
```

```python
# saving the dataframe to .csv file
data_merged.to_csv('Merged_Data.csv')
```

```python
# histograms of each column
plt.figure(figsize=(8,5))
data_merged.hist(bins=30, figsize=(12,8))
plt.tight_layout()
```

```python
# Correlation heatmap 
plt.figure(figsize=(10,8))
corr = data_merged.corr()
plt.imshow(corr, aspect='auto')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.title("Correlation Heatmap")
plt.tight_layout()
```

```python
# Income vs Home Value scatter chart. 
plt.figure(figsize=(8,6))
plt.scatter(data_merged["Median_Household_Income"], data_merged["Median_Home_Value"])
plt.xlabel("Median Household Income")
plt.ylabel("Median Home Value")
plt.title("Income vs Home Value")
plt.tight_layout()
```
